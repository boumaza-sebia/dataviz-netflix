```python
viewing_activity = pd.read_csv('../data/ViewingActivity.csv', sep=',')

def sum_duration(duration):
    return round(float(duration[0:2])*60 + float(duration[3:5]) + float(duration[6:8])/60,2)

def split_title_details(text):
    for i in range(len(text)):
        if (text[i] == ':'):
            return text[0:i], text[i+1:]
    return text, ""

#nettoyage en 4 étapes
start_date = []
start_hour = []
duration = []
title = []
details = []
for i in range(viewing_activity.shape[0]):
    temp = viewing_activity["Start Time"][i].split(' ')
    start_date.append(temp[0])
    start_hour.append(temp[1])
    
    duration.append(sum_duration(viewing_activity["Duration"][i]))
    
    temp_title, temp_details = split_title_details(viewing_activity["Title"][i])
    title.append(temp_title)
    details.append(temp_details)
    

#1. séparation de la date et de l'heure
viewing_activity.drop(columns=['Start Time'], inplace=True)
viewing_activity.insert(1, "Date", start_date, True)
viewing_activity.insert(2, "Time", start_hour, True)

#2. convertion de la durée en minutes
viewing_activity.drop(columns=['Duration'], inplace=True)
viewing_activity.insert(3, "Duration", duration, True)

#3. séparation du titre et des détails
viewing_activity.drop(columns=['Title'], inplace=True)
viewing_activity.insert(4, "Title", title, True)
viewing_activity.insert(5, "Details", details, True)

#4. suppression des données inutiles
viewing_activity.drop(columns=['Attributes'], inplace=True)
viewing_activity.drop(columns=['Supplemental Video Type'], inplace=True)
viewing_activity.drop(columns=['Bookmark'], inplace=True)
viewing_activity.drop(columns=['Latest Bookmark'], inplace=True)
viewing_activity.drop(columns=['Country'], inplace=True)


#Changement nom support
viewing_activity.replace({"Device Type": 'Netflix Linux Firefox Other'}, "Ordinateur", inplace=True)
viewing_activity.replace({'Device Type': "Netflix Windows App - Cadmium Windows Mobile"}, "Ordinateur", inplace=True)
viewing_activity.replace({'Device Type': "Apple iPhone 6s"}, "Smartphone", inplace=True)
viewing_activity.replace({'Device Type': 'Android DefaultWidevineL3Phone Android Phone'}, "Smartphone", inplace=True)
viewing_activity


viewing_activity
```

```python
def str_to_date(str_date):
    split_date = str_date.split('-')
    year = int(split_date[0])
    month = int(split_date[1])
    day = int(split_date[2])
    return datetime.datetime(year, month, day)
    
def date_to_str(date):
    res = ""
    if date.day < 10:
        res += "0"
    res += str(date.day) + "/"
    if date.month < 10:
        res += "0"
    res += str(date.month) + "/" + str(date.year)
    return res

def get_week_monday(str_date):
    temp = str_to_date(str_date)
    while((int(temp.strftime("%w"))-1)%7 != 0):
        temp = temp - datetime.timedelta(days=1)
    return temp

def get_day_number(str_date):
    split_date = str_date.split('-')
    year = int(split_date[0])
    month = int(split_date[1])
    day = int(split_date[2])
    x = datetime.datetime(year, month, day)
    return (int(x.strftime("%w"))-1)%7
```

```python
first = get_week_monday(min(viewing_activity['Date']))
last = get_week_monday(max(viewing_activity['Date']))

week_days = []
temp = first
while(last >= temp):
    week_days.append(date_to_str(temp))
    temp = temp + datetime.timedelta(days=7)

with open('../data/week_days.js', 'w') as f:
    f.write("week_days = " + str(week_days).replace('\'', '\"'))
```

```python
def get_week_number(str_date):
    week_monday = date_to_str(get_week_monday(str_date))
    for i in range(len(week_days)):
        if (week_days[i] == week_monday):
            return i
```

```python
data = {'ProfileName' : [], 'Date': [], 'Day': [] , 'Week': [], 'Title': [],'Durations': [], 'TotalDuration': []}
i = 0
while i < viewing_activity.shape[0]:
    if viewing_activity['Duration'][i] > 3:
        j = i
        str_date = viewing_activity['Date'][j]
        day = get_day_number(str_date)
        week = get_week_number(str_date)
        dict_temp = {'Profile Name' : viewing_activity['Profile Name'][j],
                    'Date' : str_date,
                    'Day': day,
                    'Week': week,
                    'Title' : [viewing_activity['Title'][j]],
                    'Durations' : [viewing_activity['Duration'][j]],
                    'TotalDuration' : 0}
        same_day = True
        while(same_day and j < viewing_activity.shape[0]):
            if (viewing_activity['Profile Name'][j] == dict_temp["Profile Name"]
                and viewing_activity['Date'][j] == dict_temp["Date"]):
                if viewing_activity['Duration'][j] > 3:
                    if viewing_activity['Title'][j] not in dict_temp["Title"]:
                            dict_temp["Title"].append(viewing_activity['Title'][j])
                            dict_temp["Durations"].append(viewing_activity['Duration'][j])
                    dict_temp["TotalDuration"] += viewing_activity['Duration'][j]
                j = j + 1
            elif (viewing_activity['Date'][j] != dict_temp["Date"]):
                same_day = False
            else:
                j = j + 1
        data["ProfileName"].append(dict_temp["Profile Name"])
        data["Date"].append(dict_temp["Date"])
        data["Day"].append(dict_temp["Day"])
        data["Week"].append(dict_temp["Week"])
        data["Title"].append(dict_temp["Title"])
        data["Durations"].append(dict_temp["Durations"])
        data["TotalDuration"].append(dict_temp["TotalDuration"])
        i = j
    else:
        i = i + 1
    
df_viewing_activity = pd.DataFrame.from_dict(data)
df_viewing_activity
```

```python
first = get_week_monday(min(viewing_activity['Date']))
last = get_week_monday(max(viewing_activity['Date'])) + datetime.timedelta(days=6)

days_list = []
temp = first
while(last >= temp):
    temp_str = str(temp.year) + "-"
    if temp.month < 10:
        temp_str += "0"
    temp_str += str(temp.month) + "-"
    if temp.day < 10:
        temp_str += "0"
    temp_str += str(temp.day)
    days_list.append(temp_str)
    temp = temp + datetime.timedelta(days=1)
    
days_list
```

```python
profiles = ["Both", "Hana", "Tarik"]

def build_empty_row(profile_, date_):
    return {"Date": date_, "ProfileName" : profile_,  
            "Day": get_day_number(date_), "Week": get_week_number(date_),
            "Title": [], "Durations": [], "TotalDuration" : 0}

dict_viewing_activity = {}

for d in days_list:
    temp = df_viewing_activity[df_viewing_activity["Date"] == d]
    dict_viewing_activity[d] = {}
    for p in profiles:
        dict_viewing_activity[d][p] = build_empty_row(p, d)
        
for d in days_list:
    temp = df_viewing_activity[df_viewing_activity["Date"] == d]
    dict_viewing_activity[d] = {}
    for p in profiles:
        dict_viewing_activity[d][p] = build_empty_row(p, d)
    
for i in df_viewing_activity.index:
    dict_viewing_activity[df_viewing_activity["Date"][i]][df_viewing_activity["ProfileName"][i]] = {"Date": df_viewing_activity["Date"][i], 
            "ProfileName" : df_viewing_activity["ProfileName"][i], "Day": df_viewing_activity["Day"][i], 
            "Week": df_viewing_activity["Week"][i], "Title": df_viewing_activity["Title"][i], 
            "Durations": df_viewing_activity["Durations"][i], "TotalDuration" : df_viewing_activity["TotalDuration"][i]}

tuples = []
for d, v in dict_viewing_activity.items():
    for p, t in v.items():
        tuples.append(t)      
```

```python
df = pd.DataFrame(columns=df_viewing_activity.columns)
for t in tuples:
    df = df.append(t, ignore_index=True)
    
df
```

```python
for i in df.index:
    df["Date"][i] = date_to_str(str_to_date(df["Date"][i]))
```

```python
result = df.to_json(orient="records")

parsed = json.loads(result)

json.dumps(parsed)

with open('../data/viewing_activity.json', 'w') as f:
    json.dump(parsed, f )
```

```python
favorites = {"Both": {}, "Hana": {}, "Tarik": {}}

for i in viewing_activity.index:
    if viewing_activity["Duration"][i] > 3:
        if viewing_activity["Title"][i] in favorites[viewing_activity["Profile Name"][i]]:
            favorites[viewing_activity["Profile Name"][i]][viewing_activity["Title"][i]] \
                += viewing_activity["Duration"][i]
        else:
            favorites[viewing_activity["Profile Name"][i]][viewing_activity["Title"][i]] \
                = viewing_activity["Duration"][i]
```

```python
tuples = []

for p, v in favorites.items():
    for t, d in v.items():
        tuples.append([p, t, round(d,2)])
        
df_favorites = pd.DataFrame(tuples,  columns =['Profile', 'Title', 'Duration'])
```

```python
with open('../data/favorites.json', 'w', encoding='utf-8') as f:
    json.dump(favorites, f, ensure_ascii=False, indent=4)
    
    
df_favorites.to_csv('../data/favorites.csv', index=False, sep=';') 
```

```python
to_remove = []
add_to_both = []
color_level = []
for i in df_favorites.index:
    for j in df_favorites.index:
        if (j > i \
            and df_favorites["Title"][i] == df_favorites["Title"][j] \
            and df_favorites["Profile"][i] != df_favorites["Profile"][j]) :
            if df_favorites["Profile"][i] != "Both":
                add_to_both.append(["Both", df_favorites["Title"][i], \
                                   max(df_favorites["Duration"][i], df_favorites["Duration"][i])])
                color_level.append((100*df_favorites["Duration"][j])/(df_favorites["Duration"][i]+df_favorites["Duration"][j]))
                if i not in to_remove:
                    to_remove.append(i)
            if j not in to_remove:
                to_remove.append(j)
            
temp = pd.concat([pd.DataFrame(add_to_both, columns=['Profile', 'Title', 'Duration'])],ignore_index=True)
temp = temp.append(df_favorites.drop(to_remove, axis=0), ignore_index=True)

for i in df_favorites.index:
    if i not in to_remove:
        if df_favorites["Profile"][i] == "Hana":
            color_level.append(0)
        elif df_favorites["Profile"][i] == "Tarik":
            color_level.append(100)
        else:
            color_level.append(50)

temp.insert(3, "ColorLevel", color_level)
temp
```

```python
temp.to_csv('../data/favorites_color_level.csv', index=False, sep=';') 
```

```python
result = temp.to_json(orient="records")

parsed = json.loads(result)

json.dumps(parsed)

with open('../data/favorites_color_level.json', 'w') as f:
    json.dump(parsed, f )
```
