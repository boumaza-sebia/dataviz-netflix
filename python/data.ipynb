```python
viewing_activity = pd.read_csv('../data/ViewingActivity.csv', sep=',')

def sum_duration(duration):
    return float(duration[0:2])*60 + float(duration[3:5]) + float(duration[6:8])/60

def split_title_details(text):
    for i in range(len(text)):
        if (text[i] == ':'):
            return text[0:i], text[i+1:]
    return text, ""

#nettoyage en 4 étapes
start_date = []
start_hour = []
duration = []
title = []
details = []
for i in range(viewing_activity.shape[0]):
    temp = viewing_activity["Start Time"][i].split(' ')
    start_date.append(temp[0])
    start_hour.append(temp[1])
    
    duration.append(sum_duration(viewing_activity["Duration"][i]))
    
    temp_title, temp_details = split_title_details(viewing_activity["Title"][i])
    title.append(temp_title)
    details.append(temp_details)
    

#1. séparation de la date et de l'heure
viewing_activity.drop(columns=['Start Time'], inplace=True)
viewing_activity.insert(1, "Date", start_date, True)
viewing_activity.insert(2, "Time", start_hour, True)

#2. convertion de la durée en minutes
viewing_activity.drop(columns=['Duration'], inplace=True)
viewing_activity.insert(3, "Duration", duration, True)

#3. séparation du titre et des détails
viewing_activity.drop(columns=['Title'], inplace=True)
viewing_activity.insert(4, "Title", title, True)
viewing_activity.insert(5, "Details", details, True)

#4. suppression des données inutiles
viewing_activity.drop(columns=['Attributes'], inplace=True)
viewing_activity.drop(columns=['Supplemental Video Type'], inplace=True)
viewing_activity.drop(columns=['Bookmark'], inplace=True)
viewing_activity.drop(columns=['Latest Bookmark'], inplace=True)
viewing_activity.drop(columns=['Country'], inplace=True)


viewing_activity
```

```python
def get_date_info(str_date):
    split_date = str_date.split('-')
    year = int(split_date[0])
    month = int(split_date[1])
    day = int(split_date[2])
    x = datetime.datetime(year, month, day)
    return (int(x.strftime("%w"))-1)%7, x.strftime("%U")
```

```python
data = {'Profile Name' : [], 'Date': [], 'Day': [] , 'Week': [], 'Title': [],'Duration': []}
i = 0
while i < viewing_activity.shape[0]:
    j = i
    date = viewing_activity['Date'][j]
    day, week = get_date_info(date)
    dict_temp = {'Profile Name' : viewing_activity['Profile Name'][j],
                'Date' : date,
                'Day': day,
                'Week': week,
                'Title' : viewing_activity['Title'][j],
                'Duration' : viewing_activity['Duration'][j]}
    same_day = True
    while(same_day and j < viewing_activity.shape[0]):
        if (viewing_activity['Profile Name'][j] == dict_temp["Profile Name"]
            and viewing_activity['Date'][j] == dict_temp["Date"]
            and viewing_activity['Title'][j] == dict_temp["Title"]):
            if (j == i):
                dict_temp["Duration"] = 0
            dict_temp["Duration"] += viewing_activity['Duration'][j]
            j = j + 1
        elif (viewing_activity['Date'][j] != dict_temp["Date"]):
            same_day = False
        else:
            j = j + 1
    data["Profile Name"].append(dict_temp["Profile Name"])
    data["Date"].append(dict_temp["Date"])
    data["Day"].append(dict_temp["Day"])
    data["Week"].append(dict_temp["Week"])
    data["Title"].append(dict_temp["Title"])
    data["Duration"].append(dict_temp["Duration"])
    i = j
    
df_viewing_activity = pd.DataFrame.from_dict(data)
df_viewing_activity
```

```python
#suppression du bruit (bandes annonce, double connexion, extraits, etc.) 
noise_index = []
for i in range(df_viewing_activity.shape[0]):
    if (df_viewing_activity["Duration"][i] < 3):
        noise_index.append(i)
        
df_viewing_activity.drop(noise_index, inplace=True)        

df_viewing_activity
```

```python
result = df_viewing_activity.to_json(orient="records")

parsed = json.loads(result)

json.dumps(parsed)

with open('../data/viewing_activity.json', 'w') as f:
    json.dump(parsed, f )
```
